```python
#print first 3 rows of summary_embeddings table
import sqlite3

conn = sqlite3.connect('GNBN.db')
cursor = conn.cursor()
# Execute the query to select the first 3 rows
cursor.execute("SELECT * FROM summary_embeddings LIMIT 3;")
rows = cursor.fetchall()

# Print the 2nd row (index 1, since indexing starts at 0)
if len(rows) > 1:
    print(rows[1])
else:
    print("The 2nd row does not exist.")

conn.close()
```
