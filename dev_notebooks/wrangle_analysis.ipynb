```python
#dependencies
import pandas as pd
import transformers
from transformers import pipeline
```

```python
df = pd.read_csv('../data/NewsArticles.csv', encoding='latin1')
display(df.head())

#drop columns after 6
df = df.drop(df.columns[6:], axis=1)
display(df.head())

#drop rows with missing values in text
display(len(df))
df = df.dropna(subset=['text'])
display(len(df))
```

```python
from transformers import pipeline

summarizer = pipeline("summarization")

df['summary'] = df['text'].apply(lambda x: summarizer(x, max_length=150, min_length=50, do_sample=False)[0]['summary_text'])

print(df[['summary']].head())
```
